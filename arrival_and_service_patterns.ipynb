# Introduction to the exponential distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

In queuing systems it is typically assumed that 

1. customers arrive at random
2. service time varies randomly.

The notion of *random* is closely associated with the **exponential distribution** (note this sometimes called the *negative* exponential distribution).  Its density function is given by (1).

\begin{equation}
f(t) = \lambda e^{-\lambda t}
\tag{1}
\end{equation}



```python
t = np.linspace(0, 10, 100)
```

```python
plt.plot(np.e**(-50 * t))
```

```python
LAMBDA = 1.0
f_t = LAMBDA * np.power(np.e, -LAMBDA * t)
```

```python
ax = pd.DataFrame(f_t, index=t).plot(figsize=(6, 5))
ax.legend(['$\lambda = 1.0$']);
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 5))

labels = []
for _lambda in range(1, 4):
    f_t = LAMBDA * np.power(np.e, -_lambda * t)
    ax.plot(pd.DataFrame(f_t, index=t))
    labels.append(f'$\lambda = {float(_lambda)}$')
            
ax.legend(labels);
```
